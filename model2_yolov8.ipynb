```python
# MS COCO Dataset으로 사전학습된 YOLO 모델 중 하나
# index : 0~79 (총 80가지 클래스)
# 모델 종류 : yolo8n, yolo8s, yolo8m, yolov8l, yolov8x
import ultralytics
from ultralytics import YOLO

# object detection + segmentation 가능한 모델
model = YOLO('yolov8x-seg.pt')
```

```python
# 예측할 object : 사람, 가방, 책, 컴퓨터
# 해당 index : 0 24 73 63
results = model.predict(source='/kaggle/input/train-img/img/*.jpg', save=True, classes=[0, 24, 73, 63], boxes=False)

# source : test 이미지가 있는 경로, 본인 환경에 맞게 수정하면 됩니다!

# predict 메소드에서 다양한 파라미터 존재
# classes : 해당 클래스만 보여지게 함
# boxes=True : bounding box를 보여지게 하여 segmentation과 더불어 예측
```
